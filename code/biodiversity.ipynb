```python
species_df = df[['species', 'decimalLatitude', 'decimalLongitude']]
```

```python
species_df.head()
```

```python

from shapely.geometry import Point

# Create a new geometry column in your pandas dataframe
geometry = [Point(xy) for xy in zip(species_df.decimalLongitude, species_df.decimalLatitude)]
geo_species_df = gpd.GeoDataFrame(species_df, geometry=geometry)
```

```python
geo_species_df.set_crs(epsg=4326, inplace=True)

gdf_3.set_crs(epsg=4326, inplace=True)
# If necessary, transform one of the dataframes to match the other's CRS
geo_species_df = geo_species_df.to_crs(gdf_3.crs)
```

```python
geo_species_df.head()
```

```python
joined = gpd.sjoin(geo_species_df, gdf_3, how="inner", op="intersects")
```

```python


counties_to_species = {}

for index, row in joined.iterrows():
    if str(row['species']) != 'nan':
        if row['NAME_2'] in counties_to_species:
            counties_to_species[row['NAME_2']].add(row['species'])

        else:
            counties_to_species[row['NAME_2']] = set()
            counties_to_species[row['NAME_2']].add(row['species'])




```

```python
print(counties_to_species)
```

```python
gdf_3 = gdf_3[['NAME_2', 'geometry']]
```

```python
gdf_3
```

```python
counties_to_species_counts = {}

for c in counties_to_species:

    counties_to_species_counts[c] = len(counties_to_species[c])
```

```python
counties_to_species_counts
```

```python
gdf_3['species_count'] = gdf_3.apply(lambda row: counties_to_species_counts.get(row['NAME_2'], 0), axis=1)
```
